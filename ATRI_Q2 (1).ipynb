```python
import numpy as np
from keras.datasets import fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train.reshape(len(X_train), 28*28) / 255.0
X_test  = X_test.reshape(len(X_test), 28*28) / 255.0
```

```python
class NeuralNet:
  def __init__(self, sizes):
    self.sizes = sizes
    self.weights = []
    self.biases = []

    for i in range(1, len(sizes)):
      W = np.random.randn(sizes[i-1], sizes[i]) * 0.1
      b = np.zeros((1, sizes[i]))
      self.weights.append(W)
      self.biases.append(b)

  def softmax(self, inputs):
    exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))
    probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
    return probabilities

  def forward(self, X):
    A = X
    total_layers = len(self.weights)

    for i in range(total_layers):
      Z = np.dot(A, self.weights[i]) + self.biases[i]
      if i == total_layers - 1:
        A = self.softmax(Z)
      else:
        A = np.maximum(0, Z)

    return A
```

```python

#hI
sizes = [784, 128, 64, 10]
model = NeuralNet(sizes)

probs = model.forward(X_train[:1])

print("Probabilities: ", probs)
```
